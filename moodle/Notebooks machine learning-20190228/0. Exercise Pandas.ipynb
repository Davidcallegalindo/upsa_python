```python
import pandas as pd
import numpy as np
from unidecode import unidecode
raw = pd.read_csv('2852sc.csv', sep=";", skiprows=6, nrows=53)
raw = raw.rename(columns={'Unnamed: 0': 'Province'})
raw = raw[~raw["Province"].str.contains("Total Nacional")]
raw
```

```python
def get_name(name):
    cods = " ".join(name.split(" ")[1:])
    return (cods.split("/")[0]).decode('utf-8').encode('utf-8')
vget_name = np.vectorize(get_name)

raw["Province"] = vget_name(raw["Province"])
provinces = raw[["Province", "1996", "2015"]]
provinces
```
